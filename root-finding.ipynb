# Root Finding Techniques

```python
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return math.e**(-x) - x
```

```python
global error, values
error = []
values = []
```

# Bi-section method

```python

def bisection(f, l, r):
    e = (r - l)
    error.append(e)
    e /= 2
    m = (l + r)/2
    values.append(m)
    if e <= 10**(-10):
        return m
    if f(l) * f(m) < 0:
        return bisection(f, l , m)
    else:
        return bisection(f, m, r)
```

```python
indices = list(range(len(error)))
plt.plot(indices, error, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Log(Error) with each iteration')
plt.yscale('log')  
plt.grid(True)
plt.show()
```

```python
plt.plot(indices, values, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Values for m')
plt.title('Value of midpoint with each iteration') 
plt.grid(True)
plt.show()
```

# Newton Raphson method

```python
error = []
values = []
```

```python
# Derivative of the function
def f_(x):
    return -math.exp(-x) - 1
```

```python
def newton_raphson(x):
    newx = x - (f(x) / f_(x))
    e = abs(newx - x)
    error.append(e)
    values.append(newx)
    if e < 10**(-10):
        return newx
    else:
        return newton_raphson(newx)
```

```python
indices = list(range(len(error)))
plt.plot(indices, error, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Log(Error) with each iteration')
plt.yscale('log')  
plt.grid(True)
plt.show()
```

```python
plt.plot(indices, values, marker='o')
plt.xlabel('Iteration')
plt.ylabel('Values for m')
plt.title('Value of midpoint with each iteration') 
plt.grid(True)
plt.show()
```
